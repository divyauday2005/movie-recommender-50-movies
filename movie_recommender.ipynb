```python
print("Total movies:", len(df))
```

```python
for i, movie in enumerate(df['title'], start=1):
    print(f"{i}. {movie}")
```

```python
cv = CountVectorizer()
genre_matrix = cv.fit_transform(df['genres'])
```

```python
similarity = cosine_similarity(genre_matrix)
```

```python
def recommend(movie_name):
    movie_name = movie_name.lower()
    df['title_lower'] = df['title'].str.lower()

    if movie_name not in df['title_lower'].values:
        print("Movie not found in dataset")
        return

    index = df[df['title_lower'] == movie_name].index[0]
    distances = similarity[index]

    movies_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )[1:6]

    print(f"\nMovies similar to '{df.iloc[index].title}':\n")
    for i in movies_list:
        print(df.iloc[i[0]].title)
```

```python
movie = input("Enter a movie name: ")
recommend(movie)
```
